```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

spotify = pd.read_csv('Spotify_data.csv')  # upload chosen dataset
```

```python
spotify['artists'] = spotify['artists'].str.strip('[\']')
```

```python
spotify.head()
```

```python
# spotify = spotify[spotify.popularity > spotify.popularity.quantile(.95)]
spotify2 = spotify[spotify.popularity > np.percentile(spotify.popularity,99.99)] # faster
```

```python
len(spotify2)
```

```python
spotify2.head()
```

```python
spotify2.to_csv('data_by_genres2.csv', index=False)
```

```python
# spotify_dup =  spotify[spotify.duplicated(['name'], keep=False)].sort_values("name")
```

```python
spotify_dup = spotify[spotify.duplicated(subset=['name', 'artists'], keep=False) & 
     ~spotify.duplicated(['name','year'], keep=False)].sort_values("name")
```

```python
# spotify.sort_values(by=['artists'], inplace=True)
# spotify_dup = spotify[spotify['artists'].duplicated(keep=False)]
```

```python
len(spotify_dup)
```

```python
spotify_dup2 = spotify_dup[spotify_dup.groupby(['name', 'artists'])['name'].transform('size') > 4]
spotify_dup2 = spotify_dup2.sort_values(['artists', 'year'], ascending = [True, True])
```
